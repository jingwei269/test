# 面向对象编程-- Object Oriented Programming

```python
### 定义类 Student，里面包含两个方法（函数）：
### 初始化__init__函数，和print_score函数

class Student(object):
    def __init__(self,name,score):
        self.name = name
        self.score = score
    def print_score(self):
        print('%s: %s' % (self.name,self.score))
    
    def get_grade(self):
        if self.score >= 90:
            return 'A'
        elif self.score >= 60:
            return 'B'
        else:
            return 'C'
```

```python
bart = Student('Bart Simpson', 59) ## 定义实例bart
lisa = Student('Lisa Simpson', 87) ## 定义实例lisa
bart.print_score()  ## 打印成绩
lisa.print_score()  
```

```python
bart.name = 'Bart Test'
bart.print_score()
```

```python
bart.get_grade()
```

```python
bart.age = 8  ### Python 允许对实例变量绑定任何数据。
bart.age
```

## 访问限制

```python
## 在Class内部，可以有属性和方法，而外部代码可以通过直接调用
## 实例变量的方法来操作数据。
## 如果要让内部属性不被外部访问，可以把属性名称前加__。

class Student(object):
    def __init__(self,name,score):
        self.__name = name
        self.__score = score
    def print_score(self):
        print('%s: %s' % (self.__name,self.__score))
    
    def get_grade(self):
        if self.__score >= 90:
            return 'A'
        elif self.__score >= 60:
            return 'B'
        else:
            return 'C'
    def get_name(self):
        return self.__name
    def get_score(self):
        return self.__score
    
    def set_name(self,name):
        self.__name = name
    
    def set_score(self,score):
        if 0 <= score <= 100:
            self.__score = score
        else:
            raise ValueError('Bad Score')
```

```python
bart = Student('Bart Simpson', 59)
```

```python
bart.get_grade()
```

```python
bart.__name  ### 外部无法访问 __name属性
```

```python
## Python解释器将对外变量__name变成了_Student__name
## 不建议这么访问变量名，不同版本的Python
## 解释器变量名可能不一致。
bart._Student__name  
```

```python
bart.get_name()  ## 通过 get方法获取属性。
```

```python
bart.set_score(90)  ## 通过set方法修改属性
bart.get_score()
```

### 在python中，变量名类似__XXX__的，是特殊变量，特殊变量是可以直接访问的，不是private变量。


# 继承和多态

```python
class Animal(object):
    def run(self):
        print('Animal is running...')
```

```python
class Dog(Animal):   ## 类Dog继承 类Animal
    pass
```

```python
adog = Dog()
adog.run()
```

```python
class Dog(Animal):   ## 类Dog继承 类Animal
    def run(self):   ## 重写主类方法
        print('Dog is running....')
    def eat(self):   ##也可增加方法
        print('Dog is eating meat ....')
bdog = Dog()
bdog.run()
```

```python
bdog.eat()
```

## 多态

```python
#### 多态，定义一个参数为类的函数：
def run_twice(animal):
    animal.run()
    animal.run()
```

```python
run_twice(Dog())
```

```python
run_twice(Animal())
```

```python
class Tortoise(Animal): ## 定义新类Tortoise
    def run(self):
        print('Tortoise is running slowly...')
```

```python
## 可以直接调用多态函数
run_twice(Tortoise()) 
```

## 静态语言 VS 动态语言
#### 如Java，如果需要传入Animal类型，则传入的对象必须是Animal类型或者其子类，否则无法调用run()方法。

#### 但python这种动态语言来说，不一定需要传入Animal类型，只需要保证传入的对象有一个run()方法就可以了。

```python
### 动态， 定义一个新类Timer不继承Animal，但有run方法
class Timer(object):
    def run(self):
        print('Start...')
```

```python
## 一样可以调用多态函数：
run_twice(Timer())
```

```python
## 操作一个对象的属性
class MyObject(object):
    def __init__(self):
        self.x = 9
    def power(self):
        return self.x * self.x
obj = MyObject()
```

```python
hasattr(obj, 'x') #  有属性‘X’吗？
```

```python
obj.x
```

```python
hasattr(obj, 'y') #  有属性‘y’吗？
```

```python
setattr(obj, 'y', 19)  #设置一个属性‘y’
```

```python
hasattr(obj, 'y') #  有属性‘y’吗？
```

```python
getattr(obj, 'y') ## 获取属性‘y'
```

```python
obj.y
```

```python
## 获取z属性，如果不存在，返回默认值404
getattr(obj, 'z', 404) 
```

```python
## 也可以获取对象方法：
hasattr(obj,'power') # 有属性power吗？
```

```python
getattr(obj,'power') ## 获取属性power
```

```python
#获取方法power并赋值到变量fn
fn = getattr(obj,'power') 
```

```python
fn()  ##调用fn()与调用obj.power()是一样的
```

```python
## 获取的是方法，则可以如下调用方法
getattr(obj,'power')()
```
